# Biến đổi dữ liệu với apply

> `DataFrame.apply(<Hàm biến đổi>)` thay đổi tất cả giá trị trong cột. Hàm biến đổi có thể là hàm có sẵn hoặc tự định nghĩa.

```python
# Hàm xử lý dữ liệu "$..,..." -> Number
def process_gdp(gdp_str):
    # Cắt "$"
    gdp_str = gdp_str[1:]
    # Cắt space ("Khoảng trắng") ở sau cùng
    gdp_str = gdp_str.strip()
    # Cắt dấu ","
    gdp_str = gdp_str.replace(",", "")
    # Ép kiểu float
    gdp_float = float(gdp_str)
    return gdp_float
process_gdp("$21,427,700,000,000 ")
```

```python
# Apply vào series
gdp_country_series_new = df["gdp_country"].apply(lambda gdp: process_gdp(gdp))
gdp_country_series_new.head()
# to_numberic -> chuyển đổi kiểu dữ liệu từ string -> số
pd.to_numeric(gdp_country_series_new).dtype
```

# Nhóm dữ liệu (groupby)

- Phân tích: Những lĩnh vực nào có tổng lượng tài sản lớn nhất?
  - Nhóm cột finalWorth (tr USD) theo cột industrials /category
  - Tính tổng cột finalWorth
  - Sắp xếp theo thứ tự giảm dần

```python
tong_cate = df[["category", "finalWorth"]].groupby("category").sum()
# DataFrame.sort_values(by=<tên cột>, ascending= True/False)
tong_cate_sorted = tong_cate.sort_values(by = "finalWorth", ascending= True)
tong_cate_sorted.head(10)
```
