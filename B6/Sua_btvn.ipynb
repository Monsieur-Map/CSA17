```python
import pandas as pd
from matplotlib import pyplot as plt #Hoặc có thể ghi import matplotlib.pylot as plt

```

```python
df = pd.read_csv("Spotify_2024_Global_Streaming_Data.csv")
df.isna().sum()
df.head()
```

```python
df.plot()
```

# plot:
- Biểu đồ đường cong

```python
# Vẽ biểu đồ đường cho Monthly Listeners (Millions)
plt.figure(figsize=(14,6))
plt.plot(df["Monthly Listeners (Millions)"], marker="^", linestyle = "-", linewidth=1, color="indigo")
plt.ylabel("Monthly Listeners (Millions)", fontsize=14)
plt.title("Biểu đồ đường số người nghe theo tháng",fontsize=20)

plt.grid(True, linestyle="--", alpha = 0.6)

#Hiển thị
plt.tight_layout()
plt.show()
```

# Bar
- Biểu đồ cột


```python
# Combine group datas (Top 10)
artist_streams = df.groupby("Artist")["Total Streams (Millions)"].sum().nlargest(10)
artist_streams
```

```python
# Vẽ biểu đồ cột ngang
plt.figure(figsize=(10,5))
artist_streams.plot(kind="barh", color="#E45A92", edgecolor="violet")

plt.title("Top 10 nghệ sĩ có lượng stream cao nhất", fontsize=20)
plt.xlabel("Total streams (mil)", fontsize=14)
plt.ylabel("Artists", fontsize=14)

#Đảo trục để nghệ sĩ cao nhất nằm ở trên
plt.gca().invert_yaxis()

plt.grid(axis="x", linestyle="--", alpha=0.6)

#Hiển thị
plt.tight_layout()
plt.show
```

# Histogram
- Biểu đồ cột đứng

```python
# add df từ billionaires
df_bili = pd.read_csv("../B4\Billionaires Statistics Dataset.csv")
df_bili.isnull().sum()

```

```python
#Xử lý null của cột tuổi -> Tính tuổi trung bình
df_bili["age"] = df_bili["age"].fillna(df_bili["age"].mean().round())
df_bili["age"]
```

```python
# Đổi kiểu int cho age
df_bili["age"] = df_bili["age"].astype(int)
df_bili["age"]
```

```python
#Vẽ biểu đồ tuổi
plt.hist(x=df_bili["age"], color="#E45A92", edgecolor="green")
plt.title("Biểu đồ cột phần tuổi của tỷ phú")
plt.ylabel("Số người")
plt.xlabel("Số tuổi")

plt.tight_layout()
plt.show()
```
